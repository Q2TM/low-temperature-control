```python
import json
import numpy as np

# Load temperature data from JSON file
with open('response18.10-18.13.json', 'r') as f:
    data = json.load(f)

# Extract kelvin temperatures from the metrics and convert to Celsius
temperatures = [metric['channels'][0]['kelvin'] - 273.15 for metric in data['metrics']]

# Convert to numpy array
temperatures = np.array(temperatures)

# Calculate mean and standard deviation
mean_temp = np.mean(temperatures)
std_temp = np.std(temperatures)

# Calculate Signal to Noise Ratio (SNR) as mean / std
snr = mean_temp / std_temp

# Calculate SNR in dB (assuming voltage ratio, so 20*log10)
snr_db = 20 * np.log10(snr)

print(f"Mean Temperature: {mean_temp:.4f} C")
print(f"Standard Deviation: {std_temp:.6f} C")
print(f"Signal to Noise Ratio: {snr:.2f}")
print(f"Signal to Noise Ratio (dB): {snr_db:.2f} dB")
print(f"Inverse Signal to Noise Ratio: {1/snr:.10f}")
```
