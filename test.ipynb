```python
import matplotlib.pyplot as plt

def graph(data):
    temps = data["temperatures"]
    resistance = data["sensors"]

    # Keep only valid (non-zero) calibration points
    x = [t - 273.15 for t, r in zip(temps, resistance) if t > 0 and r > 0]
    y = [r for t, r in zip(temps, resistance) if t > 0 and r > 0]

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, marker="o", linestyle="-")
    plt.xlabel("Temperature (°C)")
    plt.ylabel("Resistance (Ohm)")
    plt.title(f"Temperature vs Resistance (Channel {data['channel']})")
    plt.grid(True)
    plt.show()
```

```python
import numpy as np

def extrapolate_to_temp(data, target_c=-10, step_c=1, fit_points=25):
    # valid points only
    valid = [(t, r) for t, r in zip(data["temperatures"], data["sensors"]) if t > 0 and r > 0]
    temps_c = np.array([t - 273.15 for t, _ in valid], dtype=float)
    resist  = np.array([r for _, r in valid], dtype=float)

    # sort cold -> hot
    idx = np.argsort(temps_c)
    temps_c = temps_c[idx]
    resist = resist[idx]

    # fit log(R) vs T using coldest points
    n = min(fit_points, len(temps_c))
    x_fit = temps_c[:n]
    y_fit = np.log(resist[:n])
    m, b = np.polyfit(x_fit, y_fit, 1)

    # create extrapolated temperatures down to target_c
    current_min = temps_c.min()
    if target_c >= current_min:
        return data  # nothing to extrapolate

    extra_c = np.arange(current_min - step_c, target_c - step_c, -step_c)
    extra_r = np.exp(m * extra_c + b)

    # combine and return in hot -> cold order (like your original data)
    all_c = np.concatenate([temps_c, extra_c])
    all_r = np.concatenate([resist, extra_r])

    order = np.argsort(all_c)[::-1]  # hot -> cold
    all_k = all_c[order] + 273.15

    return {
        "channel": data["channel"],
        "temperatures": all_k.tolist(),
        "sensors": all_r[order].tolist(),
    }

# usage
data1_ext = extrapolate_to_temp(data1, target_c=-10, step_c=1, fit_points=25)
graph(data1_ext)
```
